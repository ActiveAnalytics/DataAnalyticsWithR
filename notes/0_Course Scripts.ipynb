## Variadic functions and ellipsis

```r
# Function with 2 variables
adder <- function(x, y)
{
    x + y
}
print(adder(3, 4))
```

```r
# Variadic function
# Negative index removes items ...
adder_v <- function(...)
{
    x <- c(...)
    result <- 0
    while(length(x) > 0)
    {
        result <- result + x[1]
        x <- x[-1]
    }
    return(result)
}
print(adder_v(1, 2, 3, 4, 5))
```

```r
# Ellipsis must be the last item
adder_v2 <- function(square = FALSE, ...)
{
    result <- 0
    x <- c(...)
    if(square)
        x <- x^2
    while(length(x) > 0)
    {
        result <- result + x[1]
        x <- x[-1]
    }
    result
}
print(adder_v2(square = TRUE, 1, 2, 3, 4, 5))
```

```r
# Ellipsis used to pass parameters from one function to another
adder_v3 <- function(message, ...)
{
    cat(message, adder_v(...), "\n")
    invisible() # returns noiselessly
}
print(adder_v3("Sum: ", 1, 2, 3, 4, 5))
```

## Linear Regression

```r
auto <- read.csv("../../data/auto/auto.csv")
auto <- auto[with(auto, price > 0 & year >= 1980),]
auto$year <- factor(auto$year)
auto[1:4,] %>% print
```

```r
table(is.na(auto$engV))
```

```r
# lm, model.frame, model.matrix, model.extract, model.response

# model.frame
mf <- model.frame(log(price) ~ body + year + mileage + engV + engType + registration,
                 data = auto, na.action = na.omit)
mf[1:4,] %>% print
```

```r
model.matrix(log(price) ~ body + year + mileage + engV + engType + registration, data = mf) -> mm
model.matrix(log(price) ~ body + year + mileage + engV + engType + registration, data = auto) -> mm
mm[1:2,] %>% print
```

```r
mr <- model.extract(mf, "response")
mr[1:4] %>% print
```

```r
# Matrix calculation for beta
system.time(coeffs <- solve(t(mm) %*% mm) %*% t(mm) %*% mr)
object.size(coeffs)
coeffs[,1] %>% print
```

## Test and train data sets (model validation basics)

```r
set.seed(0)
train_rows <- sample(nrow(dat), round(.75*nrow(dat)))
```

```r
train_rows[1:10] %>% print
```

```r
train_model <- glm(class ~ ., data = dat[train_rows,], family = quasibinomial(link = "logit"))
```

```r
mod <- predict(train_model, newdata = dat[-train_rows,], type = "response")
```

```r
((mod >= 0.5) == (dat[-train_rows, "class"] == "p")) %>% mean
```

```r
# Actual values
act <- 1*(dat[-train_rows, "class"] == "p")
pred <- 1*(mod >= 0.5)
```

```r
dat[-train_rows, "class"] %>% table
```

```r
# Number of true positives
tp <- function(actual, predicted)
{
    sum((actual == 1) & (predicted == 1))
}
```

```r
tp(act, pred) %>% print
```

```r
# Number of true negatives
tn <- function(actual, predicted)
{
    sum((actual == 0) & (predicted == 0))
}
```

```r
tn(act, pred) %>% print
```

```r
# The number of false positives
fp <- function(actual, predicted)
{
    sum((actual == 0) & (predicted == 1))
}
```

```r
fp(act, pred) %>% print
```

```r
fn <- function(actual, predicted)
{
    sum((actual == 1) & (predicted == 0))
}
```

```r
fn(act, pred) %>% print
```

```r
# True positive rate
tpr <- function(...)
{
    .tp <- tp(...); .fn <- fn(...)
    .tp/(.tp + .fn)
}
```

```r
tpr(act, pred) %>% print
```

```r
# False positive rate
fpr <- function(...)
{
    1 - tpr(...)
}
```

```r
fpr(act, pred) %>% print
```

```r
# Precision or positive predictive value
ppv <- function(...)
{
    .tp <- tp(...); .fp <- fp(...)
    .tp/(.tp + .fp)
}
```

```r
ppv(act, pred)
```

```r
# True negative rate
tnr <- function(...)
{
    .tn <- tn(...); .fp <- fp(...)
    .tn/(.tn + .fp)
}
```

```r
tnr(act, pred)
```

```r
# Negative predictive value
npv <- function(...)
{
    .tn <- tn(...); .fn <- fn(...)
    .tn/(.tn + .fn)
}
```

```r
npv(act, pred) %>% print
```
